```python
#Missing values imputation

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt   #Data visualisation libraries 
import seaborn as sns
%matplotlib inline

data = pd.read_csv('Dados.csv', sep=';',thousands=',')

data = pd.DataFrame(data, columns=data.columns)
data = data.dropna()
data.describe(include='all')
#print(data)
data.replace('NaN', 0)
from sklearn.impute import SimpleImputer

imp = SimpleImputer(strategy='constant', fill_value='NaN', missing_values=-1, copy=True)
imp.fit(data.values)
mat = imp.transform(data.values)
data = pd.DataFrame(mat, columns=data.columns)
data.describe(include='all')

y = data.pop('LOCATION').values
#print(data)

data_2007 = data[data.TIME == 2007].astype(float)
#print(data_2007.corr())
data_2007.to_csv(r'aw7.csv', index = False)
data_2008 = data[data.TIME == 2008].astype(float)
#print(data_2008.corr())
data_2008.to_csv(r'aw8.csv', index = False)
data_2009 = data[data.TIME == 2009].astype(float)
#print(data_2009.corr())
data_2009.to_csv(r'aw9.csv', index = False)
data_2010 = data[data.TIME == 2010].astype(float)
print("data_2010")
data_2010 = data_2010.corr()
data_2010.to_csv(r'aw10.csv', index = False)

data_2011 = data[data.TIME == 2011].astype(float)
print("data_2011")
data_2011 = data_2011.corr()
data_2011.to_csv(r'aw11.csv', index = False)
data_2012 = data[data.TIME == 2012].astype(float)
print("data_2012")
data_2012 = data_2012.corr()
data_2012.to_csv(r'aw12.csv', index = False)
data_2013 = data[data.TIME == 2013].astype(float)
print("data_2013")
data_2013 = data_2013.corr()
data_2013.to_csv(r'aw13.csv', index = False)
data_2014 = data[data.TIME == 2014].astype(float)
print("data_2014")
data_2014 = data_2014.corr()
data_2014.to_csv(r'aw14.csv', index = False)

data_2015 = data[data.TIME == 2015].astype(float)
print("data_2015")
data_2015 = data_2015.corr()
data_2015.to_csv(r'aw15.csv', index = False)
data_2016 = data[data.TIME == 2016].astype(float)
print("data_2016")
data_2016 = data_2016.corr()
data_2016.to_csv(r'aw16.csv', index = False)
data_2017 = data[data.TIME == 2017].astype(float)
print("data_2017")
data_2017 = data_2017.corr()
data_2017.to_csv(r'aw17.csv', index = False)
print(data_2017)

data.to_csv(r'healthyNoMissing.csv', index = False)

#print(data.head)
#data
```
